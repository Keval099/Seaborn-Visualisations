```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
dots = sns.load_dataset('dots')
```

```python
sns.catplot(x='day', y='total_bill',col='sex',jitter = False, data=tips)
plt.gcf().set_size_inches(10,4)
# jitter makes all observations into a single line
```

```python
sns.catplot(x='day', y='total_bill',kind='swarm', data=tips, col='sex', hue='smoker', row='time',height=3)
```

```python
sns.catplot(x='smoker', y='tip', data=tips)
```

```python
sns.set_style('darkgrid')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='time')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='sex', dodge =False)
# dodge makes data overlap or u can say box plot overlap
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds.sort_values('color'))
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', data=tips, col='time', row='smoker', height=4, palette="cubehelix", inner= 'stick')
# inner makes plots intrepretable in somewhat way
```

```python
# violin and swarm plots together
g = sns.catplot(x='day', y='total_bill', kind='violin', data=tips, palette="cubehelix")
sns.swarmplot(x='day', y='total_bill', data=tips, color="k",size=4, ax=g.ax)
```

```python
sns.catplot(x='sex', y='survived', data=titanic, kind='bar', hue='class', col='pclass', row='embark_town', height=3.5)
```

```python
sns.catplot(x='deck', kind='count', palette='YlOrBr', hue='sex',data=titanic, col='class', height=3)
```

```python
# point plot
sns.catplot(x='sex',y='survived', kind='point', data=titanic, hue='class')
```

```python
# strip and violin plots together
h = sns.catplot(x='embark_town',y='age', kind='strip', data=titanic, palette='magma')
sns.violinplot(x='embark_town',y='age', data=titanic, ax=h.ax, palette='light:b',jitter=False)
```
