```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('darkgrid')
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```
