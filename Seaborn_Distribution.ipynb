```python
from numpy.random import randn
```

```python
x = randn(100)
```

```python
sns.distplot(x, kde=True, bins=30, rug=True, hist=True)
```

```python
sns.kdeplot(x, shade=True)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip)
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, kind='hex', color='red')
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip, kind='kde', color='red', cbar=True)
```

```python
f, ax = plt.subplots(figsize=(6,6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(x=tips.total_bill, y=tips.tip, cmap=cmap, n_levels=60, shade=True)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
g = sns.jointplot(x=tips.total_bill, y=tips.tip, kind='kde',color='m', cmap=cmap,  n_levels=60, shade=True)

g.plot_joint(plt.scatter, c='w', s=30, linewidth=1, marker='+')
```

```python
iris= sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=20)
```
